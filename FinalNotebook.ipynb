```python
import pandas as pd
```

```python
years = [str(i) for i in range(2009, 2023)]
```

```python
pd.read_html('https://www.pgatour.com/stats/stat.02671.y2022.html')[1]
```

```python
list_dfs = []

for year in years:
    print(year)
    df = pd.read_html('https://www.pgatour.com/stats/stat.02671.y{}.html'.format(year))[1]
    df['year'] = year
    list_dfs.append(df)
```

```python
list_dfs[0]
```

```python
df = pd.concat(list_dfs)
```

```python
df.to_csv('data/golf.csv')
```
